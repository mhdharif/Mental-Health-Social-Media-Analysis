```python
import matplotlib.pyplot as plt
import pandas as pd

# Create age groups
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 20, 30, 40, 50, 60, 100],
    labels=['0-20','21-30','31-40','41-50','51-60','61+']
)

# Calculate average screen time per age group
age_screen = df.groupby('Age_Group')['Daily_Screen_Time(hrs)'].mean()

# Four-color palette
colors = ['#FF9999', '#99FF99', '#99CCFF', '#FFD700']   # red, green, blue, gold

plt.figure(figsize=(8,6))
plt.barh(
    age_screen.index,
    age_screen.values,
    color=[colors[i % 4] for i in range(len(age_screen))]   # repeating 4 colors
)

plt.xlabel("Average Daily Screen Time (hrs)")
plt.ylabel("Age Group")
plt.title("Average Daily Screen Time by Age Group")

plt.gca().invert_yaxis()  # optional: biggest bar on top
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()
```

```python
# PIE CHART

gender_screen = df.groupby('Gender')['Daily_Screen_Time(hrs)'].mean()

plt.figure(figsize=(7,7))

plt.pie(
    gender_screen,
    labels=['Male', 'Female', 'Other'],   # rename labels
    autopct='%1.1f%%',                    # show percentage
    startangle=90,                        # rotate for visual appeal
    colors=['#FF6B6B', '#4ECDC4', '#556270']   # multi-color palette
)

plt.title("Average Screen Time (Gender)")
plt.show()
```

```python
# BOX PLOT

# Create bins for screen time
df['Screen_Time_Group'] = pd.cut(
    df['Daily_Screen_Time(hrs)'],
    bins=[0, 2, 4, 6, 8, 10],
    labels=["0-2 hrs", "2-4 hrs", "4-6 hrs", "6-8 hrs", "8-10 hrs"]
)

plt.figure(figsize=(8,6))
sns.boxplot(
    x='Screen_Time_Group',
    y='Days_Without_Social_Media',
    data=df,
    palette="Set2"   # multiple colors
)

plt.xlabel("Daily Screen Time (Grouped)")
plt.ylabel("Days Without Social Media")
plt.title("Screen Time vs Days Without Social Media")
plt.xticks(rotation=30)
plt.show()
```

```python
# MULTIPLE BAR PLOT

platforms = [
    'Social_Media_Platform_Facebook',
    'Social_Media_Platform_Instagram',
    'Social_Media_Platform_LinkedIn',
    'Social_Media_Platform_TikTok',
    'Social_Media_Platform_X (Twitter)',
    'Social_Media_Platform_YouTube'
]

# Summing each platform by gender
platform_gender = df.groupby('Gender')[platforms].sum()

# Transpose so that platforms become rows and genders become columns
plot_data = platform_gender.T

# Plot multi-bar with requested colors
plot_data.plot(kind='bar', figsize=(12,6), color=['blue', 'green', 'orange'])

plt.title("Social Media Platform Usage by Gender")
plt.xlabel("Social Media Platforms")
plt.ylabel("Number of Users")
plt.legend(["Male", "Female", "Other"])
plt.xticks(
    ticks=range(len(plot_data.index)),
    labels=['Facebook','Instagram','LinkedIn','TikTok','Twitter','YouTube'],
    rotation=0)
plt.tight_layout()
plt.show()
```

```python
# LINE PLOT

# Compute average sleep quality for each gender
gender_sleep = df.groupby('Gender')['Sleep_Quality(1-10)'].mean()

# Prepare labels and values
genders = gender_sleep.index
values = gender_sleep.values

# Tri-color palette
colors = ['#FF6B6B', '#4ECDC4', '#556270']  # light red, aqua green, steel blue

plt.figure(figsize=(8,5))

# Plot line connecting points
plt.plot(
    range(len(genders)),
    values,
    color='gray',           # neutral connecting line
    linewidth=2,
    alpha=0.6
)

# Plot each point with a different color
for i in range(len(genders)):
    plt.scatter(
        i,
        values[i],
        color=colors[i],
        s=120,
        edgecolor='black',
        linewidth=1.2,
        zorder=5
    )

plt.xticks(range(len(genders)),['Male','Female','Other'])
plt.ylabel("Average Sleep Quality (1–10)")
plt.xlabel("Gender")
plt.title("Gender-wise Sleep Quality (Tri-Color Plot)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# SCATTER PLOT

# Light green + Rose colors
dual_colors = ['#90EE90', '#FF69B4']   # LightGreen, HotPink(Rose)
colors = [dual_colors[i % 2] for i in range(len(df))]

plt.figure(figsize=(7,5))
plt.scatter(
    df['Exercise_Frequency(week)'],
    df['Happiness_Index(1-10)'],
    c=colors,
    s=60,
    edgecolor='black'
)

plt.xlabel("Exercise Frequency (per week)")
plt.ylabel("Happiness Index")
plt.title("Exercise Frequency vs Happiness Index")

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# PAIR PLOT

# Rename Screen_Time_Group → Screen_Category (optional but cleaner)
df['Screen_Category'] = df['Screen_Time_Group']

# Select numeric columns + the category column
cols = ['Daily_Screen_Time(hrs)', 'Happiness_Index(1-10)', 
        'Stress_Level(1-10)', 'Screen_Category']

# Create multicolor pairplot
sns.pairplot(
    df[cols],
    hue='Screen_Category',
    palette='Set2',
    diag_kind='kde',
    height=2.5
)

plt.suptitle(" Happiness and Screen Time by Category", y=1.02)
plt.show()
```

```python
# HISTOGRAM

# Stress Level Distribution

plt.figure(figsize=(7,5))
plt.hist(df['Stress_Level(1-10)'], bins=10,edgecolor = 'white')
plt.xlabel("Stress Level (Rating:1–10)")
plt.ylabel("Count")
plt.title("Distribution of Stress Levels")
plt.show()
```

```python
# BAR PLOT

import numpy as np

# Calculate average stress level for each sleep quality
sleep_stress = df.groupby('Sleep_Quality(1-10)')['Stress_Level(1-10)'].mean()

# Create multi-color palette
colors = plt.cm.Set2(np.linspace(0, 1, len(sleep_stress)))

# Bar plot
plt.figure(figsize=(8,5))
plt.bar(
    sleep_stress.index.astype(str),
    sleep_stress.values,
    color=colors
)

plt.xlabel("Sleep Quality (Rating:1–10)")
plt.ylabel("Average Stress Level (Rating:1–10)")
plt.title("Average Stress Level Across Sleep Quality Levels")

plt.show()

```
